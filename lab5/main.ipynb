```python
import numpy as np
import pandas as pd
```

```python

c = np.asarray([1, 1, 1], dtype=np.float16) * 1/3
b = np.asarray([4, 4, 4, 4], dtype=np.float16) * 1/3
a = np.asarray([1, 1, 1], dtype=np.float16) * 1/3
d = np.asarray([5, 6, 6, 5], dtype=np.float16) * 1/3
def alg(c, b, a, d, n):
    alp = np.zeros(n, dtype=np.float16)
    bet = np.zeros(n, dtype=np.float16)
    x = np.zeros(n, dtype=np.float16)
    y = np.zeros(n, dtype=np.float16)
    y[0] = b[0]
    if b[0] == 0:
        return
    alp[0] = -c[0]/y[0]
    bet[0] = d[0]/y[0]
    for i in range(1, n-1):
        y[i] = b[i] + a[i-1]*alp[i-1]
        if y[i] == 0:
            return
        alp[i] = -c[i]/y[i]
        bet[i] = (d[i] - a[i-1]*bet[i-1])/y[i]
    y[n - 1] = b[n - 1] + a[n - 2]*alp[n - 2]
    if y[n-1] == 0:
        return
    bet[n-1] = (d[n-1] - a[n-2]*bet[n-2])/y[n-1]
    x[n-1] = bet[n-1]
    for i in range(n-2, -1, -1):
        x[i] = alp[i]*x[i+1] + bet[i]
    return x

# print(alg(c, b, a, d, 4))
```

```python
def yr(x):
    return -0.999375 + 0.999375 * np.exp(4*x) - 0.4375*x - 0.375*x**2 - 0.5*x**3
p = -4
q = 0
a = 0
b = yr(1)
def f(x):
    return 6*x**2+1
```

```python
n = 10
h = 1 / n
X = np.asarray([i*h for i in range(n+1)])
A = np.zeros(n-2)
B = np.zeros(n -1)
C = np.zeros(n-2)
D = np.zeros(n - 1)
for i in range(n-2):
    A[i] = 1 - h / 2 * p
    C[i] = 1 + h / 2 * p
for i in range(n-1):
    B[i] = h**2*q-2
for i in range(n-1):
    D[i] = h**2*f(X[i])
D[0] -= a * (1 - h / 2 * p)
D[-1] -= b * (1 + h / 2 * p)
Y = alg(C, B, A, D, len(D))
```

```python
Y = np.concatenate((np.asarray([a]), Y))
Y = np.append(Y, b)
data = pd.DataFrame({
    "x": X,
    "y": yr(X),
    "y'": Y,
    "y - y'": yr(X) - Y
})
data
```
